#Social Network Ads Data (Classification)

This data includes age and estimated salary of the user. The purchased column indicates weather the particular user with age and estimated salary have bought the product or not by viewing the social ads of the product .

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('Social_Network_Ads.csv')
print(df.shape)
print(df.head())
```

To visualize the data better I'm gonna visualize the data with age versus estimated salary and the dots as gren or blue if it was purchased or not.

```python
# import matplotlib as mpl
df0 = df[df['Purchased']==0]
df1 = df[df['Purchased']==1]
x_0 = np.asarray(df0['Age'])
x_1 = np.asarray(df1['Age'])
y_0 = np.asarray(df0['EstimatedSalary'])
y_1 = np.asarray(df1['EstimatedSalary'])


plt.style.use('seaborn-muted')
fig, ax = plt.subplots(figsize=(7, 7))
scatter1 = ax.scatter(x_0,y_0)
scatter2 = ax.scatter(x_1,y_1)
ax.set_xlabel('Age')
ax.set_ylabel('Salary')
ax.set_title('Advertisement purchase effect on employees')
ax.legend([scatter1,scatter2],['Not purchased', 'Purchased'])
plt.show()
```

You can clearly see that young people with a salary lower than 100000 are very likely to not purchase things if they seen an advertisement. We could make a classification model that separte our data very clearly, for example with an SVM Classification model.
